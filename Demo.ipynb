```python
import json
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from multi_label_classification import plot_binary_confusion_matrices, plot_multi2single_confusion_matrx
```

# Demo - Multi Label Classification

## Data - Imagine a multi label classification task contains 4 animal.

```python
with open("sampledata.json", "r") as f:
    d = json.loads(json.load(f))
y_true = np.array(d['True'])
y_pred = np.array(d['Pred'])
class_names = ["Ape", "Bird", "Cat", "Dog"]
print(f"Shape of y_true: {y_true.shape}, y_pred: {y_pred.shape}")
```

## binary_confusion_matrices

```python
fig, axes = plt.subplots(2, 2, figsize=(12,10))
cms = plot_binary_confusion_matrices(y_true, y_pred, class_names, fig, axes, cbar=True)
```

## multi2single_confusion_matrx

```python
plt.figure(figsize=(18,15))
cm = plot_multi2single_confusion_matrx(y_true, y_pred, class_names, normalize=None, cbar=True)
```
